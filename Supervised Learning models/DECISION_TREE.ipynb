```python
#DECISION TREE
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

db=datasets.load_breast_cancer()
X_train, X_test, Y_train, Y_test = train_test_split(db.data,db.target,test_size = 0.2,random_state=0)
dt = DecisionTreeClassifier(random_state = 10)
dt.fit(X_train, Y_train)
dt.score(X_test, Y_test)

Y_pred=dt.predict(X_test)

print(metrics.accuracy_score(Y_test,Y_pred))
print(metrics.confusion_matrix(Y_test,Y_pred))
print(db.target_names)

fig = plt.figure(figsize=(25,20))
tree.plot_tree(dt, 
                   feature_names=db.feature_names,  
                   class_names=db.target_names,
                   filled=True)
```

```python
#DECISION TREE FOR LOAD_IRIS
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

db = load_iris()
x_train , x_test, y_train, y_test = train_test_split(db.data,db.target,test_size = 0.2,random_state = 0)
knn = DecisionTreeClassifier(random_state = 10)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)

y_pre = knn.predict(x_test)
print(metrics.accuracy_score(y_test,y_pre))
print(metrics.confusion_matrix(y_test,y_pre))
tree.plot_tree(knn)
```
