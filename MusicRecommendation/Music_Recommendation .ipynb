# Music Recommendation System 


# Dataset information

Million song dataset contain two files : triplet files and metadata file.triplet file contain user_id, song_id, listern time.
The metadatafile contains song_id, release, year, title and artist_name.Million song Dataset is a mixture of  song from various
website with the rating that users gave after listening the songs.

There are three type of recommendation system : content-based, collaborative and popularity.
    content-based:- use keywords for recommendation.
    collaborative:- if one user hear two common songs and second user hear another third song then it will suggest to first 
                    user to hear that song.
    popularity:- Based on the stars and review of public it will suggest the song.

# Import Datasets

```python
import pandas as pd
import numpy as np
import recommender as recommender
```

# Loading the datasets

```python
song_df_1 = pd.read_csv('triplets_file.csv')
song_df_1.head()
#head()-used to return top n rows of dataframes or series.by default n=5.
```

```python
song_df_2 = pd.read_csv('song_data.csv')
song_df_2.head()
```

# Data Preprocessing

```python
#merging two dataframe with removing duplicates from second frame
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on='song_id',how = 'left')
song_df.head()
```

```python
#finding length of dataframe
len(song_df)
```

```python
#merging to frame that are title and artist name and make one frame named as song
song_df['song'] = song_df['title']+'-'+song_df['artist_name']
song_df.head()
```

```python
#finding the cummulative sum of listen count of the songs
song_grouped = song_df.groupby(['song']).agg({'listen_count' : 'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = (song_grouped['listen_count']/grouped_sum) * 100
song_grouped.sort_values(['listen_count','song'],ascending = [0,1])
#here listern count will be sort by descending values bcz we gave 0 but when two values are same it will sort that in ascending order with the help of 1 written over there.
```

# Popularity Recommendations model

```python
pr = recommender.popularity_recommender_py()
pr.create(song_df,'user_id','song')
#using this create method from recommender.py file we can able to create the recommendation system
```

```python
pr.recommend(song_df['user_id'][5])
#will recommend the song for user of 5th num..although it won't affect by user it will show top 10 songs or trending songs from score or rank
```

# Item Similarity Recommendations 

```python
ir = recommender.item_similarity_recommender_py()
ir.create(song_df,'user_id','song')
```

```python
#print the song listen by index num 10th user
user_items = ir.get_user_items(song_df['user_id'][5])
for user_item in user_items:
    print(user_item)
```
